```python
import random

lista = []
def criaLista(N):
    global lista
    lista = []
    for i in range(N):
        lista.append(random.randint(1,100))
    return lista
def leLista(N):
    global lista
    lista = []
    for i in range(N):
        lista.append(int(input(f"introduza o número {i+1}/{N}")))
    return lista
def somaLista(lista):
    soma = 0
    for n in lista:
        soma = n + soma
    return soma
def mediaLista(lista):
    res = 0
    for n in lista:
        res = n + res
    return res / len(lista)
def maiorLista(lista):
    maior = lista[0]
    for elem in lista:
        if elem > maior:
            maior = elem
    return maior
def menorLista(lista):
    menor = lista[0]
    for elem in lista:
        if elem < menor:
            menor = elem
    return menor
def estaOrdenadaC(N):
    for i in range(len(N)-1):
        if N[i] > N[i+1]:
            return "Não"
    return "Sim"
def estaOrdenadaD(N):
    for n in range(len(N)-1):
        if N[n] < N[n+1]:
            return "Não"
    return "Sim"
def indiceDe(N,elem):
    condição = -1
    if elem in N:
        condição = N.index(elem) + 1
    return condição

escolha = -1

while escolha != 0:
    escolha = int(input("Escolha um modo (1-Criar Lista, 2-Ler Lista, 3-Soma Lista, 4-Média, 5-Maior, 6-Menor, 7-esta ordenada por ordem crescente, 8-esta ordenada por ordem decrescente, 9-Procura um elemento, 0-Sair):"))
    if escolha == 1:
        print(criaLista(int(input("Quantos elementos deseja ter na lista?"))))
    elif escolha == 2:
        print(leLista(int(input("Quantos números deseja inserir?"))))
    elif escolha == 3:
        print(somaLista(lista))
    elif escolha == 4:
        print(mediaLista(lista))
    elif escolha == 5:
        print(maiorLista(lista))
    elif escolha == 6:
        print(menorLista(lista))
    elif escolha == 7:
        resultado = estaOrdenadaC(lista)
        print(resultado)
    elif escolha == 8:
        resultado = estaOrdenadaD(lista)
        print(resultado)
    elif escolha == 9:
        print(lista)
        print(indiceDe(lista, int(input("Que número desejas procurar?"))))
if escolha == 0:
    print(lista, "O jogo terminou.")
```
