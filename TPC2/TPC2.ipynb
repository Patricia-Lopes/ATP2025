```python
import random


fósforos = 21
soma = int(fósforos)
print("Jogo dos fósforos, o número inicial é 21. Cada jogador pode tirar entre 1 a 4 fósforos. Quem tirar o último perde.")

resposta = input("Desejas começar? (Sim ou Não): ")

if resposta == "Sim":
    while soma > 1:
        n = int(input("Quantos fósforos deseja tirar?"))
        if n == 1 or n == 2 or n == 3 or n == 4:
            soma -= n
            print(f"Tiraste {n} fósforos. Restam {soma} fósforos.")
        else:
            print("Número inválido, escolha um número entre 1 e 4.")
            continue
        if soma == 1:
            print("Ganhaste!")
            break

        computador = min(5 - n, soma - 1)
        soma -= computador
        print(f"O computador tirou {computador} fósforos. Restam {soma} fósforos.")
        if soma == 1:
            print("Perdeste!")
            break

elif resposta == "Não":
    
    computador = random.randint (1, 4)
    soma -= computador
    print(f"O computador tirou {computador} fósforos. Restam {soma} fósforos.")

    while soma > 1:
        n = int(input("Quantos fósforos deseja tirar?"))
        if n == 1 or n == 2 or n == 3 or n == 4:
            soma -= n
            print(f"Tirante {n} fósforos. Restam {soma} fósforos.")
        else:
            print("Número inválido, escolha um número entre 1 e 4.")
            continue
        if soma == 1:
            print("Ganhaste!")
            break

        if soma % 5 != 1:
            computador = (soma -1) % 5
            if computador == 0:
                computador = random.randint(1, min(4, soma - 1))
        else:
            computador = random.randint(1, min(4, soma - 1))
        soma -= computador 
        print(f"O computador tirou {computador} fósforos. Restam {soma} fósforos.")
        if soma == 1:
            print("Perdeste!")
            break
else:
    print("Resposta inválida")

```
